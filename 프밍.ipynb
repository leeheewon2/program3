```python
import pandas as pd
import numpy as np
```

```python
life=pd.read_csv('Life Expectancy Data.csv')
life
```

```python
# Life Expectancy라는 데이터 세트입니다. 
# 193개국의 기대수명, 건강요인과 관련된 데이터 세트는 동일한 WHO 데이터 저장소 웹사이트에서 수집되었습니다.
# 이 데이터 세트는 193개 국가에 대한 2000-2015년의 기록을 서로 다른 국가에 기반을 둔 관측치입니다. 
# 면역 요인, 사망률 요인, 경제적 요인, 사회적 요인 및 기타 건강 관련 요인에 초점을 맞췄습니다.

#변수들은 
#성별의 성인 사망률(인구 1000명당 15~60년 사망 확률), 인구 1000명당 유아 사망자 수, 알코올 1인당 소비량(15+) 기록(순수 알코올 리터 단위),
# 1인당 국내총생산(%) 대비 보건비 지출, 1세 아동 중 B형 간염(HepB) 예방접종 보장률(%), 홍역(인구 1000명당 보고된 환자 수),자원의 소득구성에 관한 인적자원개발지수(0~1)
# 학교수업년수(년),1인당 국내총생산(USD), BMI 등으로 구성돼있습니다.

#국가별 인구의 기대 수명을 파악해서 높고, 낮은 가치에 기여하는 요인을 파악하고
# 또 기대 수명을 효율적으로 향상시키기 위해 어느 지역이 중요한지 분석하고 변수에 따른 기대수명을 분석할 것입니다.
```

```python
life1=life.iloc[:,[0,1,3,4,5,6,7,10,16,21]]
life1.head()
```

```python
life_simple={'Life expectancy':'Expectancy','Adult Mortality':'Adult_Mortality','infant deaths':'infant_death','percentage expenditure':'Expenditure'}
```

```python
life2=life1.rename(columns=life_simple)
life2
```

```python
life2.isna().sum(axis=0)
```

```python
#결측치 개수를 확인한 결과
```

```python
life3=life2.dropna() #결측치 제거
```

```python
life3.assign(Expectancy=life3.Expectancy.astype(np.float16),
             Adult_Mortality=life3.Adult_Mortality.astype(np.float16),
             Alcohol=life3.Alcohol.astype(np.float16),
             BMI=life3.BMI.astype(np.float16),
             Schooling=life3.Schooling.astype(np.float16),
             Year=life3.Year.astype(np.int16),
             infant_death=life3.infant_death.astype(np.int16)).info(memory_usage='deep')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
corr = life3[['Expectancy', 'Adult_Mortality','GDP', 'Schooling','infant_death','BMI','Alcohol','Expenditure']].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(
 corr,
 mask=mask,
 fmt='.2f',
 annot=True,
 ax=ax,
 cmap='RdBu',
 vmin=-1,
 vmax=1,
 square=True)
#상관관계를 히트맵에 시각화 보고싶은 변수 색조합 red blue 
```
